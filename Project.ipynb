```python
import pandas as pd
df1 = pd.read_csv("debt_per_capita - debt_per_capita.csv")
df2 = pd.read_csv("deliquent_percentage - deliquent_percentage.csv")
df3 = pd.read_csv("ExpenditureByStudent - Sheet1.csv")
```

```python
df1_final = df1.drop(df1.columns[[1, 2, 3, 4, 5, 6, 7, 8, 15]], axis=1)
df1_final.head()
```

```python
df2_final = df2.drop(df2.columns[[1, 2, 3, 4, 5, 6, 7, 8, 15]], axis=1)
df2_final.head()
```

```python
df3.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
alternative_merge = pd.merge(
    df1_final, df2_final, on="state", suffixes=("_Debt Per Capita", "_Deliquent Percentage"))
alternative_merge.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
alternative_merge1 = pd.merge(
    alternative_merge, df3, on="state")
alternative_merge1.head()
```
